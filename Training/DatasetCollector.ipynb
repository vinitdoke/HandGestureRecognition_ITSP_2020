## Dataset Collector

* Press a to get BGSubtractor
* Press s to start recording
* Press e to end recording
* First make a dataset folder and make 7 more folder inside it with name as given "up" "down" "left" "right" "stationary1" "stationary2" "select"
* Enter your_name
* For stationary1 let palm-side fist face camera, and for 2 use backside fist

```python
gesture_name = "sample" #"up" "down" "left" "right" "stationary1" "stationary2" "select"
your_name = "Video" 

import cv2
import numpy as np

def getMask(roi, fgbg):
    fgmask=fgbg.apply(roi,learningRate=0)
    kernel=np.ones((5,5),np.uint8)
    morphopen=cv2.morphologyEx(fgmask,cv2.MORPH_OPEN,kernel)
    morphclose=cv2.morphologyEx(morphopen,cv2.MORPH_CLOSE,kernel)
    ret,thresh=cv2.threshold(morphclose,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    return thresh

def main():
    
    cap=cv2.VideoCapture(0)
    mfound=False
    one=False
    i=0
    rec=False

    while True:
        ret, frame = cap.read()
        roi=frame[150:400,450:620]
        cv2.rectangle(frame, (450,150), (620,400), (255,255,255))

        if mfound:
            mask=getMask(roi, fgbg)
            cv2.imshow("mask", mask)
            
            if rec:
            #Change file paths accordingly
                cv2.imwrite(fr"E:\USERS\VINIT\Desktop\MainProject\dataset\Video\{gesture_name}\mask_{i}.png", mask)
                cv2.imwrite(fr"E:\USERS\VINIT\Desktop\MainProject\dataset\Video\{gesture_name}\image_{i}.png", roi)
                i+=1
        cv2.imshow("frame", frame)

        inp=cv2.waitKey(1)
        if inp==27:
            cap.release()
            cv2.destroyAllWindows()
            break
        elif inp==ord('a'):
            fgbg=cv2.createBackgroundSubtractorMOG2(0,50)
            mfound=True
        elif inp==ord("s"):
            rec=True
        elif inp==ord("e"):
            rec=False

main()
```
